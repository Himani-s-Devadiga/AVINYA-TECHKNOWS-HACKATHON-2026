```python
#  GNN River Modeling + Color-coded Flood Alerts
import networkx as nx
import matplotlib.pyplot as plt
import random


G = nx.DiGraph()

nodes = ['A','B','C','D','E']
edges = [('A','B'),('B','C'),('C','D'),('D','E')]

G.add_nodes_from(nodes)
G.add_edges_from(edges)


node_levels = {n: random.randint(50, 500) for n in nodes}


node_colors = []
for n in nodes:
    if node_levels[n] < 200:
        node_colors.append('green')   # safe
    elif node_levels[n] < 350:
        node_colors.append('yellow')  # warning
    else:
        node_colors.append('red')     # danger

plt.figure(figsize=(8,5))
pos = nx.spring_layout(G, seed=42)  # layout
nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=1000, font_weight='bold', arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels={e: f"{node_levels[e[1]]} cm" for e in edges})
plt.title("GNN River Network - Flood Alert (Color-coded)")
plt.show()


print("Simulated Water Levels at Nodes:")
for n, level in node_levels.items():
    print(f"Node {n}: {level} cm")
```

```python
#  Flood Momentum Index + 48h Forecast (simulated)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
n = 20
data = pd.DataFrame({
    'rainfall_mm': np.random.randint(50, 200, n),
    'river_level_cm': np.random.randint(100, 500, n)
})


data['rainfall_mm'] = pd.to_numeric(data['rainfall_mm'], errors='coerce')
data['river_level_cm'] = pd.to_numeric(data['river_level_cm'], errors='coerce')


data['flood_momentum'] = (data['river_level_cm'] * np.random.uniform(0.5, 1.5, len(data))) + (data['rainfall_mm'] * 0.3)
data['flood_momentum'] = pd.to_numeric(data['flood_momentum'], errors='coerce')  # ensure numeric

plt.figure(figsize=(8,5))
plt.plot(range(1,n+1), data['flood_momentum'], marker='o', label='Flood Momentum')
plt.title("Flood Momentum Index (Simulated 48h Forecast)")
plt.xlabel("Time Step (hour)")
plt.ylabel("Momentum Score")
plt.xticks(range(1,n+1))
plt.grid(True)
plt.legend()
plt.show()
```

```python
#  Simulated River Network for Evacuation Routes
G = nx.DiGraph()
nodes = ['A','B','C','D','E','F','G']
edges = [('A','B'),('B','C'),('C','D'),('D','E'),('B','F'),('F','G'),('G','E')]
G.add_nodes_from(nodes)
G.add_edges_from(edges)

node_risk = {n: random.randint(20,90) for n in nodes}

node_congestion = {n: random.uniform(0,1) for n in nodes}

# Evacuation route simulation: shortest path weighted by flood risk + congestion
for u,v in G.edges():
    G[u][v]['weight'] = node_risk[v] + int(node_congestion[v]*50)  # simple heuristic

start_node = 'A'
safe_zone = 'E'
path = nx.shortest_path(G, source=start_node, target=safe_zone, weight='weight')

print("Optimal Evacuation Path from A to E considering risk & congestion:")
print(" -> ".join(path))
```

```python
# Color-code nodes based on risk
import networkx as nx
import random
import matplotlib.pyplot as plt

random.seed(42)

G = nx.DiGraph()
nodes = ['A','B','C','D','E','F','G']
edges = [('A','B'),('B','C'),('C','D'),('D','E'),('B','F'),('F','G'),('G','E')]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

node_risk = {n: random.randint(20,90) for n in nodes}

node_congestion = {n: random.uniform(0,1) for n in nodes}

for u,v in G.edges():
    G[u][v]['weight'] = node_risk[v] + int(node_congestion[v]*50)  # simple heuristic

# Evacuation simulation
start_node = 'A'
safe_zone = 'E'
path = nx.shortest_path(G, source=start_node, target=safe_zone, weight='weight')

print("Optimal Evacuation Path from A to E considering risk & congestion:")
print(" -> ".join(path))

# --- Optional Visualization ---
plt.figure(figsize=(8,5))
pos = nx.spring_layout(G, seed=42)
node_colors = [node_risk[n] for n in G.nodes()]
nx.draw(G, pos, with_labels=True, node_color=node_colors, cmap=plt.cm.Reds, node_size=800, font_weight='bold')
nx.draw_networkx_edges(G, pos, edgelist=list(zip(path,path[1:])), width=3, edge_color='blue')  # highlight path
plt.title("Evacuation Route Simulation (Blue = Optimal Path)")
sm = plt.cm.ScalarMappable(cmap=plt.cm.Reds, norm=plt.Normalize(vmin=min(node_risk.values()), vmax=max(node_risk.values())))
sm.set_array([])
plt.colorbar(sm, label='Flood Risk')
plt.show()
```

```python
# Climate Drift Detection
# Simulate historical rainfall (past 20 years)
import numpy as np
np.random.seed(42)

years = np.arange(2002, 2022)
historical_rainfall = np.random.normal(120, 30, len(years)) 

current_rainfall = 220  

mean_hist = np.mean(historical_rainfall)
std_hist = np.std(historical_rainfall)
z_score = (current_rainfall - mean_hist)/std_hist

# Display results
print("\nClimate Drift Detection:")
print(f"Current rainfall: {current_rainfall} mm")
print(f"Historical mean: {mean_hist:.2f} mm, std: {std_hist:.2f} mm")
print(f"Z-score: {z_score:.2f}")

if abs(z_score) > 2:
    print("⚠️ Significant anomaly detected (Climate Drift)")
else:
    print("Rainfall within historical range")
```
